```python
# Split the Data into Training and Testing

# Dividing the dataset in training and testing set
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
from sklearn import metrics 

np.random.seed(0)
df_train,df_test=train_test_split(master_wine,train_size=0.7,test_size=0.3,random_state=100)

```

```python
# Create our features
X = master_wine[list(master_wine.columns)[:-1]]

# Create our target
y = master_wine['quality']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=1, stratify=y)
```

```python
# Pairplot: see both distribution of single variables and relationships between two variables
sns.pairplot(df_train) 
plt.show()
```

```python
#Rescaling
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df_train.head()
```

```python
df_train[:]=scaler.fit_transform(df_train[:])
df_train.head()
```

```python
# Multi-Variate Analysis
sns.set(rc={'figure.figsize':(11.7,8.27)})
cData_attr = df_train.iloc[:, 0:7]
sns.pairplot(cData_attr, diag_kind='kde')   
```

```python
#Linear Regression

# fitting linear regression to training data
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
LinearRegression()
```

```python
train_pred = regressor.predict(X_train)
print(train_pred)
test_pred = regressor.predict(X_test) 
print(test_pred)
```

```python
# calculating Root Mean Squared Error (RMSE)
from sklearn.metrics import mean_squared_error

train_rmse = mean_squared_error(train_pred, y_train) ** 0.5
print(train_rmse)
test_rmse = mean_squared_error(test_pred, y_test) ** 0.5
print(test_rmse)
```

```python
# rounding off the predicted values for test set
predicted_data = np.round_(test_pred)
print(predicted_data)
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, test_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, test_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, test_pred)))
```
