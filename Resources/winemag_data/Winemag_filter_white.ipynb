```python
import pandas as pd
import glob
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a loop to join 250K multiple of wine enthusiast mag 

master_df = pd.DataFrame()

for file in os.listdir(os.getcwd()):
    if file.endswith('.csv'): 
        master_df = master_df.append(pd.read_csv(file))
```

```python
master_df.to_csv('Master_wine.csv', index=False)
```

```python
master_df
```

```python
#drop colummns 

master_df = master_df.drop(['description', 'designation','subregion',"subsubregion","title",'url','winery'], axis = 1)
```

```python
master_df
```

```python
#Filter by Country // We only need Portugal

master_df = master_df.query("country == 'Portugal'")
```

```python
master_df
```

```python
master_df = master_df.query("category == 'White'")
```

```python
#drop NA values for alcohol & rating

white_winedf = master_df.dropna(subset=['alcohol', 'rating', 'price'])
```
